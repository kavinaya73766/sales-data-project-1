```python
file_path="/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv"
```

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(file_path)
```

```python
data.head()
```

```python
titanic_df = pd.DataFrame(data)

age_series = titanic_df['Age']
fare_series = titanic_df['Fare']
survived_series = titanic_df['Survived']

print(age_series.head())
print(fare_series.head())
print(survived_series.head())
```

```python
print(titanic_df.head())
print(titanic_df.info())
print(titanic_df.describe())
```

```python
print(titanic_df.isnull().sum())

titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)

titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

titanic_df.dropna(inplace=True)

print("Missing values handled")
```

```python
titanic_df.reset_index(drop=True, inplace=True)
print(titanic_df.loc[0])

print(titanic_df.iloc[0])

print(titanic_df[['Age', 'Fare', 'Survived']].head())
```

```python
titanic_df.rename(columns={'Sex': 'Gender'}, inplace=True)

print(titanic_df.head())

pivot_table = titanic_df.pivot_table(values='Fare', index='Gender', columns='Pclass')
print(pivot_table)
melted_df = pd.melt(titanic_df, id_vars=['PassengerId'], value_vars=['Age','Fare'])
print(melted_df.head())
```

```python
age_above_30 = titanic_df[titanic_df['Age'] > 30]
fare_above_50 = titanic_df[titanic_df['Fare'] > 50]
female_survived = titanic_df[(titanic_df['Gender'] == 'female') & (titanic_df['Survived'] == 1)]

print(female_survived.head())

print(titanic_df.sort_values(by='Fare').head())
print(titanic_df.sort_values(by='Age').head())
print(titanic_df.sort_values(by='Survived').head())
```

```python
survival_by_gender = titanic_df.groupby('Gender')['Survived'].mean()
print(survival_by_gender)

avg_age_class = titanic_df.groupby('Pclass')['Age'].mean()
print(avg_age_class)

survival_by_embarked = titanic_df.groupby('Embarked')['Survived'].sum()
print(survival_by_embarked)
```

```python
class_info = pd.DataFrame({
    'Pclass': [1,2,3],
    'Class_Type': ['First Class','Second Class','Third Class']
})

merged_df = pd.merge(titanic_df, class_info, on='Pclass')
print(merged_df.head())

joined_df = titanic_df.join(class_info.set_index('Pclass'), on='Pclass')
print(joined_df.head())

df_part1 = titanic_df.iloc[:100]
df_part2 = titanic_df.iloc[100:200]

concat_df = pd.concat([df_part1, df_part2])
print(concat_df.head())
```

```python
avg_age_survival = titanic_df.groupby('Survived')['Age'].mean()
print(avg_age_survival)

highest_fare = titanic_df['Fare'].max()
lowest_fare = titanic_df['Fare'].min()
print("Highest Fare:", highest_fare)
print("Lowest Fare:", lowest_fare)

survival_percentage = titanic_df['Survived'].mean() * 100
print("Survival Percentage:", survival_percentage)
```
